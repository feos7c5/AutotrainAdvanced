# Image Semantic Segmentation using AutoTrain Advanced

In this notebook, we will train an image semantic segmentation model using AutoTrain Advanced.
You can replace the model with any Hugging Face transformers compatible model and dataset with any other dataset in proper formatting.
For dataset formatting, please take a look at [docs](https://huggingface.co/docs/autotrain/index).

````python
from autotrain.trainers.image_semantic_segmentation.params import ImageSemanticSegmentationParams
from autotrain.project import AutoTrainProject
````

````python
HF_USERNAME = "your_huggingface_username"
HF_TOKEN = "your_huggingface_write_token" # get it from https://huggingface.co/settings/token
# It is recommended to use secrets or environment variables to store your HF_TOKEN
# your token is required if push_to_hub is set to True or if you are accessing a gated model/dataset
````

````python
params = ImageSemanticSegmentationParams(
    model="nvidia/segformer-b0-finetuned-ade-512-512",
    data_path="scene_parse_150", # path to the dataset on huggingface hub
    image_column="image", # the column in the dataset that contains the image
    target_column="annotation", # the column in the dataset that contains the segmentation masks
    train_split="train",
    valid_split="validation",
    epochs=3,
    batch_size=4,
    lr=6e-5,
    optimizer="adamw_torch",
    scheduler="linear",
    gradient_accumulation=2,
    mixed_precision="fp16",
    project_name="autotrain-image-semantic-segmentation",
    log="tensorboard",
    push_to_hub=True,
    username=HF_USERNAME,
    token=HF_TOKEN,
)
# tip: you can use `?ImageSemanticSegmentationParams` to see the full list of allowed parameters
````

If your dataset is in CSV / JSONL format (JSONL is most preferred) and is stored locally, make the following changes to `params`:

```python

```

````python
# this will train the model locally
project = AutoTrainProject(params=params, backend="local", process=True)
project.create()
````
